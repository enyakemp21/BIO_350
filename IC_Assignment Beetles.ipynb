```python
import numpy as np
```

```python
#create an algorithm that calculates the beetle population, and the total population over 50 years
#the initial first year population is 500 and the initial second year population in 500
#on average 1 beetle produces 1 offspring in second year and 5 offspring in third year
#the population of the second year is o.3 times the second population
#the population of the third year is 0.5 times of the second year population


def calculate_beetle_population(years, initial_first_year_population, initial_second_year_population):
    """
    Calculates the beetle population over a number of years across three age classes using a transition matrix.

    Parameters:
        years (int): Number of years to calculate the population for.
        initial_first_year_population (int): Initial population of first-year beetles.
        initial_second_year_population (int): Initial population of second-year beetles.

    Returns:
        first_year_population (list): Population of first-year beetles over the years.
        second_year_population (list): Population of second-year beetles over the years.
        third_year_population (list): Population of third-year beetles over the years.
        total_population (list): Total beetle population over the years.
    """

    #complete the algorithm using matrix multiplication
    #create a matrix to represent the transition of populations between years
    A = np.array([[0, 1, 5], [0.3, 0, 0], [0, 0.5, 0]])
    #initialize the popuation vector with the initial populations
    initial_population_vector = np.array([initial_first_year_population, initial_second_year_population, 0])
    #crate lists to store the populations of each age group and total population each year
    first_year_population = []
    second_year_population = [] 
    third_year_population = []
    total_population = []
    #append the current populations to the lists
    first_year_population.append(initial_population_vector[0])
    second_year_population.append(initial_population_vector[1])
    third_year_population.append(initial_population_vector[2])
    total_population.append(initial_population_vector.sum())
    #iterate over the number of years to calculate the population for each year
    current_population_vector = initial_population_vector
    for year in range(1, years):
        #matrix multiplication to get the population for the next year
        current_population_vector = np.dot(A, current_population_vector)
        #append the populations to the lists
        first_year_population.append(current_population_vector[0])
        second_year_population.append(current_population_vector[1])
        third_year_population.append(current_population_vector[2])
        total_population.append(current_population_vector.sum())

    return first_year_population, second_year_population, third_year_population, total_population
    

A = np.array([[0, 1, 5], [0.3, 0, 0], [0, 0.5, 0]])
```

```python
#create a matrix to store the population over 50 years
n = np.zeros((3, 50))
```

```python
n[:, 0] = n_t   #set the first column to the initial population
for t in range(1, 50):
    n[:, t] = np.dot(A, n[:, t-1])   #calculate the population for each year
n.sum(axis=0)   #calculate the total population for each year   
```

```python
import matplotlib.pyplot as plt

pop_by_age = n.T  # Shape becomes (years, age_classes)


#transpose the matrix n to have years as rows and age classes as columns
years_list = range (len(pop_by_age))

#plot each age class and the total population
plt.plot(years_list, pop_by_age[:,0], label='Age 1')
plt.plot(years_list, pop_by_age[:,1], label='Age 2')
plt.plot(years_list, pop_by_age[:,2], label='Age 3')

plt.plot(years_list, pop_by_age.sum(axis=1), label='Total', color='black', linewidth=2)
plt.legend()
plt.plot(n.sum(axis=0))
plt.ylabel('Population')
plt.xlabel('Year') 
plt.xlabel('Year')
plt.ylabel('Total Population')
plt.title('Beetle Population Over 50 Years')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
