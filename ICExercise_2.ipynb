```python
#import "Population_Data.csv" into pandas dataframe
import pandas as pd
```

```python
#import "Population_Data.csv" into pandas dataframe
pop_data = pd.read_csv("Population_Data.csv")
pop_data.head()
```

```python
#average the population from days 17 to days 60
average_pop = pop_data[(pop_data['Days'] >= 17) & (pop_data['Days'] <= 60)]['Population'].mean()
print(f"Average population from days 17 to 60: {average_pop}")
```

```python
import matplotlib.pyplot as plt

plt.plot(pop_data['Days'], pop_data['Population'], marker='o', color='black', linestyle='-')
plt.axhline(y=276, color='green', linestyle='--', linewidth=2)
plt.xlabel('Days')
plt.ylabel('Population Size')
plt.title('Population Growth Over Time')
plt.xlim(left=0)
plt.ylim(bottom=0)
plt.grid(True)

# Set custom y-ticks to include 276
yticks = plt.yticks()[0]
if 276 not in yticks:
    plt.yticks(list(yticks) + [276])

plt.show()
```
