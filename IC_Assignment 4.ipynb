Chapter 4: Lotka-Volterra Competition
- Algorithm to calculate 2 seperate species to visualise interspecific competition


```python
def NandM_population(N0, M0, time, KN, KM, dN, dM, S):
    # N0: initial population size of species N
    # M0: initial population size of species M
    # time: number of time steps to simulate
    # KN: carrying capacity of species N
    # KM: carrying capacity of species M
    # dN: death rate of species N
    # dM: death rate of species M
    # S: competition coefficient (effect of species M on species N)
    # growth model
    N_values = [N0]
    M_values = [M0]  # Initialize M_values
    current_N = N0
    current_M = M0
    for _ in range(time):
        current_N = current_N + S * current_N * ((1 - current_N / KN) - (current_M / KM)) - dN * current_N
        current_M = current_M + S * current_M * ((1 - current_M / KM) - (current_N / KN)) - dM * current_M
        # round to 3 decimal places
        current_N = round(current_N, 3)
        current_M = round(current_M, 3)
        N_values.append(current_N)
        M_values.append(current_M)
    return N_values, M_values
```

```python
#Run the function with 40 days and N0 = 5, M0 = 12, time = 1, Kn = 400, KM = 1600, dN =  0.6, dM = 0.4, S = 1
result = NandM_population(12, 5, 40, 400, 1600, 0.4, 0.6, 1)    
print(result)
```

```python
import matplotlib.pyplot as plt
```

```python
# Unpack the results
N_values, M_values = NandM_population(5, 12, 40, 1600, 400, 0.6, 0.4, 1)
# Create a list of days for the x-axis

days = list(range(len(N_values)))  # Both lists have the same length
# Plot the results

plt.plot(days, N_values, marker='o', label='Species N')
plt.plot(days, M_values, marker='s', label='Species M')
# Add a title and labels
plt.title('Population Growth Over Time')
plt.xlabel('Days')
plt.ylabel('Population Size')
plt.legend()
plt.grid()
plt.show()
```

The figure above shows the population growth for 2 competing species (N and M), over a period of 40 times. 
- x-axis = time
- y-axis = population size

Species N starts with a small population of 5, it then increases slighlty before decreasing again to near 0.
Species M starts with a slightly larger population of 12, it increases significantly before stabalizing. 

This shows that species M dominates the environment. This is competitve exclusion (the inevitable elimination from a habitat of one of two different species with identical needs for resources)

Alpha MN = KM/KN = 4
- Alpha is the competition coefficient which measures the effect of species N on species M. A value of 4 means that one individual of N is equivalent to 4 individuals of M. 

Alpha NM = KN/KM = 0.25
- This is the competition coefficient for the effect of species M on species N. A value of 0.25 means that one individual of M is equivalent to 0.25 individuals of N in terms of competition.

rM = SM - dM
- This is the net growth rate of species M, calculated as its intrinsic growth rate (SM) minus its death rate (dM).

rN = SN - dN 
- This is the net growth rate of species N, calculated as its intrinsic growth rate (SN) minus its death rate (dN).

Coefficients and rates are essentially just describing how strongly one species competes with another and how quickly those species populations are affected. 
[High alpha values mean that there is strong competition]
