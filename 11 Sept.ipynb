```python
def N__logistic(N_0, r, k, time):
    # Logistic growth model
    N_values = [N_0]
    current_N = N_0
    for _ in range(time):
        # Calculate next population size using the logistic equation
        current_N = current_N + r * current_N * (1 - current_N / k)
        # Round to 3 decimal places
        current_N = round(current_N, 3)
        N_values.append(current_N)
    return N_values
```

```python
#Run the function with 365 days and r =0.6 and k = 400 with an initial population of 0.12
result = N__logistic(12, 0.6, 400, 365)
print(result)
```

```python
import matplotlib.pyplot as plt
```

```python
# plot the results with days on the x-axis and population size on the y-axis


days = list(range(366)) # 0 to 365 days
plt.plot(days, result, marker='o')
plt.title('Population Growth Over Time')
plt.xlabel('Days')
plt.ylabel('Population Size')
plt.grid()
plt.show()
```

```python
def N__logistic(N_0, s, k, d, time):
    # Logistic growth model
    N_values = [N_0]
    current_N = N_0
    for _ in range(time):
        # Calculate next population size using the logistic equation
        current_N = current_N + s * current_N * (1 - current_N / k) - d * current_N
        # Round to 3 decimal places
        current_N = round(current_N, 3)
        N_values.append(current_N)
    return N_values
```

```python
#Run the function with 35 days and r =0.6 and k = 400 with an initial population of 0.12
result = N__logistic(12, 1, 400, 0.4, 35)
print(result)
```

```python
import matplotlib.pyplot as plt
```

```python
# plot the results with days on the x-axis and population size on the y-axis

days = list(range(len(result)))  # Match the length of result
plt.plot(days, result, marker='o')
plt.title('Population Growth Over Time')
plt.xlabel('Days')
plt.ylabel('Population Size')
plt.grid()
plt.show()
```
