```python
import matplotlib.pyplot
```

```python
def N__exp(N_0 ,p , time): 
    # Exponential growth model
    # # Save N_values into a list 
    N_values = [N_0]
    # Save the initial population size
    current_N = N_0
    for _ in range(time):
        current_N = current_N + p * current_N
        # round to 3 decimal places
        current_N = round(current_N, 3)
        N_values.append(current_N)
    return N_values
```

```python
#Run the function with 40 days and p =0.2 with an initial population of 0.01
result = N__exp(0.01, 0.4, 35)
print(result)
```

```python
days = list(range(36)) # 0 to 35 days
plt.plot(days, result, marker='o')
plt.title('Population Growth Over Time')
plt.xlabel('Days')
plt.ylabel('Population Size')
plt.grid()
plt.show()
```

```python
#Change N_Exp to N_exp2 and have 3 conditions based on N_0
#If N_0 < 500, p = 0.4
#If N_0 >= 500 and N_0 < 1000, p = 0.2
#If N_0 >= 1000, p = 0
def N__exp2(N_0 , time): 
    # Exponential growth model with conditions on p
    if N_0 < 500:
        p = 0.4
    elif 500 <= N_0 < 1000:
        p = 0.2
    else:
        p = 0.0
    
    # Save N_values into a list 
    N_values = [N_0]
    # Save the initial population size
    current_N = N_0
    for _ in range(time):
        current_N = current_N + p * current_N
        # round to 3 decimal places
        current_N = round(current_N, 3)
        N_values.append(current_N)
    return N_values  


    
```

```python
#Run the function with 35 days and p =0.2 with an initial population of 0.01
result = N__exp(0.01, 35)
print(result)
```
