```python
def NandM_population(N0, M0, time, KN, KM, dN, dM, S):
    # N0: initial population size of species N
    # M0: initial population size of species M
    # time: number of time steps to simulate
    # KN: carrying capacity of species N
    # KM: carrying capacity of species M
    # dN: death rate of species N
    # dM: death rate of species M
    # S: competition coefficient (effect of species M on species N)
    # growth model
    N_values = [N0]
    M_values = [M0]  # Initialize M_values
    current_N = N0
    current_M = M0
    for i in range(time):
        if i == 9:
            current_N = current_N + 100
        current_N = current_N + S * current_N * ((1 - current_N / KN) - (current_M / KM)) - dN * current_N
        current_M = current_M + S * current_M * ((1 - current_M / KM) - (current_N / KN)) - dM * current_M
        # round to 3 decimal places
        current_N = round(current_N, 3)
        current_M = round(current_M, 3)
        N_values.append(current_N)
        M_values.append(current_M)
    return N_values, M_values
```

```python
def NandM_population(N0, M0, time, KN, KM, rN, rM, alpha, beta):
    # N0: initial population size of species N
    # M0: initial population size of species M
    # time: number of time steps to simulate
    # KN: carrying capacity of species N
    # KM: carrying capacity of species M
    # rN: intrinsic growth rate of species N
    # rM: intrinsic growth rate of species M
    # growth model
    N_values = [N0]
    M_values = [M0]  # Initialize M_values
    current_N = N0
    current_M = M0
    for i in range(time):
        if i == 14:
            current_N = current_N + 100
        current_N = current_N + rN * current_N * (1 - ((current_N + alpha * current_M) / KN))
        current_M = current_M + rM * current_M * (1 - ((current_M + beta * current_N) / KM))
        # round to 3 decimal places
        current_N = round(current_N, 3)
        current_M = round(current_M, 3)
        N_values.append(current_N)
        M_values.append(current_M)
    return N_values, M_values
```
